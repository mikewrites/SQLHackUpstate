```python
trucks = snow['truck_name'].unique().tolist()
```

```python
trucks = [int(item) for item in trucks]
```

```python
truck_info = pd.DataFrame(columns = ['year', 'make', 'model', 'milage'], index = trucks)
```

```python
truck_info
```

```python
models = [['dodge','ram'],['ford', 'f150'], ['ford', 'f250']]
for truck in truck_info.index:
    mod = models[random.randint(0,2)]
    truck_info.loc[truck, 'year'] = random.randint(1998,2021)
    truck_info.loc[truck, 'milage'] = random.randint(30000,250000)
    truck_info.loc[truck, 'make'] = mod[0]
    truck_info.loc[truck, 'model'] = mod[1]
```
